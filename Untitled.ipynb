```python
def ismatch(x,y):
    if pd.isnull(x):
        return 1
    else:
        try:
            return x.lower() == y.lower()
        except:
            return x==y
```

```python
dense_col_cn =['客户代码','客户等级','客户属性', '客户地区', '客户领域',['需求品牌','采购报价品牌'],['需求型号','采购报价型号'],['需求包装','采购报价包装'],'ECCN','采购代码','处理方式', '是否重点询价','采购报价质检','供应商类型','供应商等级','客户新老','供应商代码','供应商类型', '供应商等级',]
dense_col_en = ['CardCode','U_CusLevel','U_CusGroupCode','U_Region','U_DomainName',['Brand','U_QuoBrand'],['Modle','U_QuoModle'],['Pacakage','U_QuoPackage'],'Eccn','U_Buyer','Status','KeyPoint','U_QCDesc','U_QuoGroupCode','U_QuoLevel','newclient','U_CardCode','U_QuoGroupCode','U_QuoLevel']
```

```python
all_dict = {}
def m(data,name1,name2):
    def map_(x,dit):
        return dit[x]
    def Replenish(x):
        if pd.isnull(x):
            return "null"
        else:
            return x
    if isinstance(name1,str):
        data[name1] = data[name1].apply(Replenish)
        t = set(data[name1].tolist())
        dit = dict()
        iter_ = 0
        for i in t:
            dit[i] = iter_
            iter_ += 1
        data[name2] = data[name1].apply(map_,args=(dit,))
        all_dict[name2] = dit
    else:
        name_cn_1,name_cn_2 = name1
        name_en_1,name_en_2 = name2
        data[name_cn_1] = data[name_cn_1].apply(Replenish)
        data[name_cn_2] = data[name_cn_2].apply(Replenish)
        t = set(data[name_cn_1].tolist()) | set(data[name_cn_2].tolist()) 
        dit = dict()
        iter_ = 0
        for i in t:
            dit[i] = iter_
            iter_ += 1
        data[name_en_1] = data[name_cn_1].apply(map_,args=(dit,))
        data[name_en_2] = data[name_cn_2].apply(map_,args=(dit,))
        all_dict[name_en_1] = dit
```

```python
for x,y in zip(dense_col_cn,dense_col_en):
    print(x)
    m(data,x,y)
```

```python
import numpy as np
def helper(name,name1):
    print(name,name1)
    tmp = data[name].unique().tolist()
    bins = []
    quantiles = [10,25,50,75,90]
    while -1 in tmp:
        bins = [0]
        tmp.remove(-1)
    for i in quantiles:
        bins.append(np.percentile(tmp, i))
    def hel(x,bins):
        for name,value in enumerate(bins):
            if x < value:
                return name
        return len(bins)
    all_dict[name] = bins
    data[name1] = data[name].apply(hel,args=(bins,))
```
